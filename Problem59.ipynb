```python
class Solution:
    def generateMatrix(self, n: int):
        num = 1
        output = []
        # Create and initialize the array
        for i in range (0, n):
            tmp = []
            for j in range (0, n):
                tmp.append(None)
            output.append(tmp)
        # Start to traverse the array in spiral order
        # Use cur_pos to implement the current slot
        # Use nxt_pos to detect the boundary and decide the direction
        cur_pos = [0, 0]
        nxt_pos = [0, 1]
        direction = 1 # 1 is right, 2 is down, 3 is left, 4 is up
        while num < n*n:
            # Detect direction
            if nxt_pos[0] < n and nxt_pos[1] < n and output[nxt_pos[0]][nxt_pos[1]] == None:
                # Don't change direction 
                output[cur_pos[0]][cur_pos[1]] = num
                num += 1
                print("1", cur_pos, nxt_pos)
                cur_pos = nxt_pos.copy()
                print("2", cur_pos)
                print("3", output)
                print("dir", direction)
#                 if direction == 1:
#                     nxt_pos[1] += 1
#                 elif direction == 2:
#                     nxt_pos[0] += 1
#                 elif direction == 3:
#                     nxt_pos[1] -= 1
#                 elif direction == 4:
#                     nxt_pos[0] -= 1
                if direction == 1:
                    nxt_pos[0] = cur_pos[0]
                    nxt_pos[1] = cur_pos[1] + 1
                elif direction == 2:
                    nxt_pos[0] = cur_pos[0] + 1
                    nxt_pos[1] = cur_pos[1]
                elif direction == 3:
                    nxt_pos[0] = cur_pos[0]
                    nxt_pos[1] = cur_pos[1] - 1
                elif direction == 4:
                    nxt_pos[0] = cur_pos[0] - 1
                    nxt_pos[1] = cur_pos[1]
                print(cur_pos, nxt_pos, "dir", direction)
                
            else: # Change direction

                print("change direction", cur_pos, nxt_pos)
                if direction == 4:
                    print("4 to 1")
                    direction = 1
                else:
                    print("+1")
                    direction += 1
                
#                 if direction == 1:
#                     nxt_pos[1] += 1
#                 elif direction == 2:
#                     nxt_pos[0] += 1
#                 elif direction == 3:
#                     nxt_pos[1] -= 1
#                 elif direction == 4:
#                     nxt_pos[0] -= 1
                if direction == 1:
                    nxt_pos[0] = cur_pos[0]
                    nxt_pos[1] = cur_pos[1] + 1
                elif direction == 2:
                    nxt_pos[0] = cur_pos[0] + 1
                    nxt_pos[1] = cur_pos[1]
                elif direction == 3:
                    nxt_pos[0] = cur_pos[0]
                    nxt_pos[1] = cur_pos[1] - 1
                elif direction == 4:
                    nxt_pos[0] = cur_pos[0] - 1
                    nxt_pos[1] = cur_pos[1]
                    
        if num == n*n:
            output[cur_pos[0]][cur_pos[1]] = num
        
        return output

```
